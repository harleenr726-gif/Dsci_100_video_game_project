Data Description
1) There are two datasets: players.csv and sessions.csv.
- player.csv: 196 observations/rows and 7 variables  
- sessions.csv: 1535 observations/rows and 5 variables

2) Variables in players.csv

experience
- Type: character
- Description: player’s self-reported experience level 

subscribe
- Type: logical
- Description: whether the player subscribed to the game newsletter (TRUE/FALSE)

hashedEmail
- Type: character
- Description: hashed unique identifier used to protect the player's actual email address

played_hours
- Type: double
- Description: Total number of hours the player has spent playing

name
- Type: character
- Description: Player’s first name 

Age
- Type: double
- Description: player's age in years

gender
- Type: character
- Description: player's gender (Male or Female)

3) Variables in sessions.csv

hashedEmail
- Type: character
- Description: Hashed identifier linking each session to a player

start_time
- Type: character 
- Description: Date and time when the game session began

end_time
- Type: character 
- Description: Date and time when the game session ended

original_start_time
- Type: double
- Description: Server record of original session start time

original_end_time
- Type: double
- Description: server recorded original session end time

6. Data issues

Players data:
- Some players have 0 played_hours, which affects averages and may indicate incomplete sessions
- Experience and gender are stored as character strings but should be categorical factors
- played hours should be stored as integer and not double

Sessions data:
- start_time and end_time are character timestamps and must be converted to datetime, or values for date and time in seperate columns
- Should include timezone
- original_start_time and original_end_time are large numeric Unix timestamps needing conversion
- Some sessions appear extremely short or long, suggesting inconsistent play or logging issues

Invisible issues
- self report bias in age, gender, and experience

Data collection
- player demographics were likely collected through signup/onboarding

```r
players_plot1 <- player_data |>
    ggplot(aes(x = experience, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(title = "Subscription Rate by Experience Level",
        x = "Experience level",
        y = "Proportion of Players",
        fill = "Subscribed") +
    scale_y_continuous(labels = scales::percent)
players_plot1
```

Subscription proportions vary across experience levels. pro and regular players appear to subscribe at slightly higher rates than amateur or beginner players. Also, experience level may contain useful signal for predicting newsletter subscription.

```r
players_plot2 <- player_data |>
  ggplot(aes(x = Age, y = played_hours)) +
  geom_point(alpha = 0.5) +
  labs(title = "Age vs Total Hours Played",
       x = "age",
       y = "played time in hours")
players_plot2
```

Most players, across all ages, have very low total playtime. A few players have extremely high playtime (150 to 220+ hours), which creates clear outliers. There is no strong visible relationship between age and played_hours. The distribution is heavily right-skewed due to a small number of highly active players.

Methods and Plan

To answer my question, I will use KNN classification.

Why this method is appropriate:
KNN classification is a non-parametric method that does not assume a specific functional relationship between predictors and the response. Because player characteristics such as age, experience, and played_hours may influence subscription in nonlinear ways, KNN can capture complex patterns that simpler linear models might miss

Assumptions
- The choice of k (number of neighbors) affects model performance
- Observations that are close in the predictor space are expected to have similar outcomes
- Numeric variables must be on comparable scales because KNN relies on distance calculations

Potential limitations
- KNN can perform poorly if predictors are not scaled, especially since played_hours varies widely across players.
- The method is sensitive to noise, outliers, and irrelevant predictors.

Model comparison and selection 
- Although KNN is my primary method, I will compare its performance to linear regression used as a classifier.
- Models will be compared using: Accuracy on the test set and Multiple values of k for KNN to identify the best-performing model. The model with the best predictive performance will be selected.

Data processing plan
- Convert experience to a factor and ensure numeric variables (Age, played_hours) are correctly typed.
- Split the data into 80% training and 20% testing before fitting any models.
- Scale numeric predictors for KNN classification to ensure balanced distance calculations.
- Fit KNN models with several values of k on the training data.
- Fit a linear regression classifier for comparison.
- Evaluate both models on the test set and select the best-performing classifier.
